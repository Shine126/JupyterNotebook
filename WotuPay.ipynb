```python
import matplotlib.pyplot as plt
import numpy as np

# 
def LineforProfit(ProIn):
    # x,1ISV20
    x = np.linspace(0, ProIn['PFC'],10)  
    # CusPay
    CusPay = x/ProIn['IR']*(1+ProIn['CR'])*(1-ProIn['GR']) 
    # GovPay
    GovPay = x/ProIn['IR']*(1+ProIn['CR'])*ProIn['GR']
    # Profit 
    Profit = x - CusPay - ProIn['PC']

    # 
    plt.figure(figsize=(12, 4))
    # --Profit
    plt.plot(x, Profit, label='Profit', color='blue', linestyle='-')

    # --GovPay
    plt.plot(x, GovPay, label='GovPay', color='yellow', linestyle='--')

    # 
    plt.title('Wotu is Better!')
    plt.xlabel('Product Cost')
    plt.ylabel('Profit-blue and GovPay-yellow')

    for i in range(len(x)):
        plt.annotate(f'({x[i]:.2f}, {Profit[i]:.2f})', (x[i], Profit[i]), textcoords="offset points", xytext=(0,10), ha='center')
        plt.annotate(f'({x[i]:.2f}, {GovPay[i]:.2f})', (x[i], GovPay[i]), textcoords="offset points", xytext=(0,10), ha='center')
    # 
    plt.legend()
    plt.grid(True,linestyle='--',color='gray')
    # 
    plt.show()

######################
# CustomerPay  >>>>>>>>> CusPay
# GovermentPay >>>>>>> GovPay
# Profit >>>>>>>>>>>>>> Pr
#######################
# ISVProductCost >>>>>>>>> PC
# GovermentRate  >>>>> GR
# ISVProductForCustomer >> PFC
# ISVISVRate >>>>>>>> IR
# CloudRate >>>>>>>>>>> CR

ProItem = {
    'PC' : 0.5,
    'GR' : 0.6,
    'PFC' : 15,
    'IR' : 0.6,
    'CR' : 0.1
}

LineforProfit(ProItem)
```
