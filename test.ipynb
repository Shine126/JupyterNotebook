```python
import matplotlib.pyplot as plt
import numpy as np

# 生成x值范围
x = np.linspace(0, 1500, 10)  # 生成0到2000之间的20个均匀分布的点

# 计算对应的y值
yh = x * (1 - 0.06) + 100 - 20
yz = x * 0.99
y = yh - yz
# 创建折线图
plt.plot(x, y, label='y')

# 添加标题和标签
plt.title('line1')
plt.xlabel('x')
plt.ylabel('y')

for i in range(len(x)):
    plt.annotate(f'({x[i]:.2f}, {y[i]:.2f})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
# 添加图例
plt.legend()
plt.grid(True,linestyle='--')
# 显示图形
plt.show()
```
